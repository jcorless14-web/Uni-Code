```python
import numpy as np
import matplotlib.pyplot as plt
seq = np.zeros(20)
seq[0] = 1
seq[1] = 1
seq[2] = 1
for x in range (3,20):
    seq[x] = seq[x-1] + seq[x-2] + seq[x-3]
for x in range (20):
    print('When N =', x+1, 'the value in the sequence equals:', seq[x])
# I did it first just like the factorial example in the notes, but it read wrong so i messed around with the print function and got this which is neater
# Now the plot numbers vs place in sequence:
n = np.arange(1, 21)
plt.plot(n, seq)
plt.title("Graph of numbers vs Place in Sequence")
plt.xlabel("Place in Sequence")
plt.ylabel("Numbers")
plt.grid()
plt.show()
#When i did this first, i thought numpy was bugged due to arrange not working, i used an extra r, then i had to switch my axis' also
```
