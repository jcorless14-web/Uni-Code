```python
# The following is code which once given the image distance and the focal length of a lens, will return firstly the image distance and then the magnification
```

```python
f = eval(input("Please enter the Focal length of the lens (in cm)!"))
u = eval(input("Please enter the objects distance from the lens (in cm)!"))
v = ((1/f - (1/u))**(-1))
m = v/u
print("The Image distance equals", round(v, 1), "cm")
print("The magnification equals", round(m, 2))
```
